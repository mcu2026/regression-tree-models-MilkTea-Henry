```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 讀取資料
boston房價資料

altavish/boston-housing-dataset@kaggle

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("altavish/boston-housing-dataset")

# download dataset
df = pd.read_csv(f"{path}/HousingData.csv")
df.head()
```

## 資料探索

```python
df.describe()
```

## 檢查空值

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['CHAS']=df['CHAS'].astype('object')
```

```python
df.info()
```

```python
df[['CRIM','ZN','INDUS','AGE','LSTAT']] = df[['CRIM','ZN','INDUS','AGE','LSTAT']].fillna((df[['CRIM','ZN','INDUS','AGE','LSTAT']].mean()))
```

```python
df['CHAS']=df['CHAS'].fillna(method='bfill')
```

## 建立訓練資料

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

### 分割資料

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

## 建立CART回歸樹模型

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=6)
DT_reg=regressor.fit(X_train, Y_train)
```

```python
DT_reg
```

```python
print(DT_reg.score(X_train,Y_train))
print(DT_reg.score(X_test,Y_test))
```

```python
Y_pred=DT_reg.predict(X_test)
```

```python
Y_pred
```
